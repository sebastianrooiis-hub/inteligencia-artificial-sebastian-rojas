# Proyecto de Estadística Descriptiva
Este notebook explica las principales medidas de estadística descriptiva y sus derivados: **Media, Mediana, Moda, Varianza, Desviación, Rango y Correlación**.  
Cada sección incluye:
- Cálculos con ejemplos
- Gráficas
- Contexto explicativo


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random
```

```python
# Datos de ejemplo
data_media = [10, 14, 15, 20, 18, 22, 30, 25]
data_mediana = [5, 7, 12, 14, 18, 21, 30]
data_moda = [3, 3, 4, 4, 4, 6, 7, 8]
data_var = [10, 12, 14, 16, 18, 20]
data_rango = [2, 5, 9, 12, 20, 22]
data_x = [1, 2, 3, 4, 5, 6, 7]
data_y = [2, 4, 5, 7, 10, 11, 14]

# Contextos aleatorios
contextos = {
    "media": "Ventas diarias de una tienda; la media indica el promedio de ventas.",
    "mediana": "Puntajes de un examen; la mediana indica el valor central del grupo.",
    "moda": "Número de productos comprados; la moda muestra el valor más común.",
    "varianza": "Edades de los participantes; la varianza y desviación muestran dispersión.",
    "rango": "Temperatura máxima y mínima diaria; el rango muestra la amplitud del clima.",
    "correlacion": "Horas de estudio vs calificaciones; la correlación indica si más horas mejoran el desempeño."
}
```

## Media
La **media** representa el valor promedio de un conjunto de datos.  
Contexto: Ventas diarias de una tienda, donde queremos conocer el promedio de ventas.


```python
media = np.mean(data_media)
print(f"Media = {media}")

plt.figure(figsize=(6,4))
plt.plot(data_media, marker='o', color='blue')
plt.title(f"Media = {media}")
plt.xlabel("Día")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()
```

## Mediana
La **mediana** es el valor central de un conjunto de datos ordenado.  
Contexto: Puntajes de un examen; nos indica el estudiante promedio.


```python
mediana = np.median(data_mediana)
print(f"Mediana = {mediana}")

plt.figure(figsize=(6,4))
plt.plot(data_mediana, marker='o', color='green')
plt.title(f"Mediana = {mediana}")
plt.xlabel("Estudiante")
plt.ylabel("Puntaje")
plt.grid(True)
plt.show()
```

## Moda
La **moda** es el valor que más se repite en un conjunto de datos.  
Contexto: Número de productos comprados; indica el valor más frecuente.


```python
moda = pd.Series(data_moda).mode()[0]
print(f"Moda = {moda}")

plt.figure(figsize=(6,4))
plt.bar(range(len(data_moda)), data_moda, color='orange')
plt.title(f"Moda = {moda}")
plt.xlabel("Índice")
plt.ylabel("Cantidad")
plt.grid(True)
plt.show()
```

## Varianza y Desviación
La **varianza** mide la dispersión de los datos respecto a la media,  
y la **desviación estándar** es la raíz cuadrada de la varianza.  
Contexto: Edades de los participantes; muestra qué tan dispersas están.


```python
varianza = np.var(data_var, ddof=1)
desviacion = np.sqrt(varianza)
print(f"Varianza = {varianza}, Desviación = {desviacion}")

plt.figure(figsize=(6,4))
plt.plot(data_var, marker='o', color='red')
plt.title(f"Varianza={varianza:.2f}, Desviación={desviacion:.2f}")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```

## Rango
El **rango** es la diferencia entre el valor máximo y mínimo.  
Contexto: Temperatura diaria; indica la amplitud del clima.


```python
rango = max(data_rango) - min(data_rango)
print(f"Rango = {rango}")

plt.figure(figsize=(6,4))
plt.plot(data_rango, marker='o', color='purple')
plt.title(f"Rango = {rango}")
plt.xlabel("Día")
plt.ylabel("Temperatura")
plt.grid(True)
plt.show()
```

## Correlación
La **correlación** indica la relación entre dos variables.  
Contexto: Horas de estudio vs calificaciones; muestra si más horas mejoran el desempeño.


```python
correlacion = np.corrcoef(data_x, data_y)[0][1]
print(f"Correlación = {correlacion}")

plt.figure(figsize=(6,4))
plt.scatter(data_x, data_y, color='brown')
plt.title(f"Correlación = {correlacion:.2f}")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación")
plt.grid(True)
plt.show()
```
