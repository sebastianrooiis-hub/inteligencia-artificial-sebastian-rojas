```python
# ── 1.2 Últimas 10 filas ─────────────────────────────────────────────────────
print('=' * 60)
print('📋 ÚLTIMAS 10 FILAS')
print('=' * 60)
display(df.tail(10))
```

```python
# ── 1.3 Tipo de dato de cada columna ─────────────────────────────────────────
print('=' * 60)
print('🔍 TIPOS DE DATOS POR COLUMNA')
print('=' * 60)
print(df.dtypes)
print()
print('💡 Interpretación:')
print('   • object  → texto (cadena de caracteres)')
print('   • int64   → número entero de 64 bits')
print('   • float64 → número decimal de 64 bits')
```

## 📌 Punto 2 — Operaciones Matemáticas entre Columnas

```python
# ── 2.1 Total_Venta = Precio × Cantidad ──────────────────────────────────────
df['Total_Venta'] = df['Precio'] * df['Cantidad']

# ── 2.2 Costo_Total = Total_Venta + Costo_Envio ───────────────────────────────
df['Costo_Total'] = df['Total_Venta'] + df['Costo_Envio']

print('✅ Columnas Total_Venta y Costo_Total creadas')
display(df[['Producto', 'Precio', 'Cantidad', 'Costo_Envio', 'Total_Venta', 'Costo_Total']].head(10))
```

```python
# ── 2.3 Promedio de Total_Venta ───────────────────────────────────────────────
promedio_tv = df['Total_Venta'].mean()
print('=' * 60)
print('📊 PROMEDIO DE TOTAL_VENTA')
print('=' * 60)
print(f'   → Promedio: $ {promedio_tv:,.0f}')
print()
print('💡 En promedio, cada producto genera')
print(f'   $ {promedio_tv:,.0f} en ventas totales.')
```

```python
# ── 2.4 Producto con MAYOR Total_Venta ───────────────────────────────────────
idx_max      = df['Total_Venta'].idxmax()
producto_max = df.loc[idx_max, 'Producto']
valor_max    = df.loc[idx_max, 'Total_Venta']

# ── 2.5 Producto con MENOR Total_Venta ───────────────────────────────────────
idx_min      = df['Total_Venta'].idxmin()
producto_min = df.loc[idx_min, 'Producto']
valor_min    = df.loc[idx_min, 'Total_Venta']

print('=' * 60)
print('🏆 PRODUCTO CON MAYOR TOTAL_VENTA')
print('=' * 60)
print(f'   → Producto : {producto_max}')
print(f'   → Total    : $ {valor_max:,.0f}')
print()
print('=' * 60)
print('📉 PRODUCTO CON MENOR TOTAL_VENTA')
print('=' * 60)
print(f'   → Producto : {producto_min}')
print(f'   → Total    : $ {valor_min:,.0f}')
```

---
## 📌 Punto 3 — Porcentajes y Análisis

```python
# ── 3.1 IVA = 19% del Total_Venta ────────────────────────────────────────────
df['IVA'] = df['Total_Venta'] * 0.19

# ── 3.2 Ganancia_Estimada = 25% del Total_Venta ───────────────────────────────
df['Ganancia_Estimada'] = df['Total_Venta'] * 0.25

# ── 3.3 Perdida_Estimada = 0.05% del Total_Venta ─────────────────────────────
df['Perdida_Estimada'] = df['Total_Venta'] * 0.0005

print('✅ Columnas de porcentajes creadas')
display(df[['Producto', 'Total_Venta', 'IVA', 'Ganancia_Estimada', 'Perdida_Estimada']].head(10))
```

```python
# ── 3.4 Porcentaje de cada producto sobre el total general ────────────────────
total_general = df['Total_Venta'].sum()
df['Porcentaje_Ventas'] = (df['Total_Venta'] / total_general) * 100

print('=' * 60)
print('📊 PORCENTAJE DE VENTAS POR PRODUCTO (Top 15)')
print('=' * 60)
print(f'   Total general de ventas: $ {total_general:,.0f}')
print()
display(
    df[['Producto', 'Total_Venta', 'Porcentaje_Ventas']]
    .sort_values('Porcentaje_Ventas', ascending=False)
    .head(15)
    .assign(Porcentaje_Ventas=lambda x: x['Porcentaje_Ventas'].map('{:.2f}%'.format))
    .reset_index(drop=True)
)
```

```python
# ── 3.5 Productos con participación > 15% ─────────────────────────────────────
productos_top = df[df['Porcentaje_Ventas'] > 15].copy()

print('=' * 60)
print('🔝 PRODUCTOS CON PARTICIPACIÓN > 15% EN VENTAS')
print('=' * 60)

if productos_top.empty:
    print('   ⚠️  Ningún producto supera el 15% del total de ventas.')
    print('   Esto indica que las ventas están bien distribuidas.')
    print()
    print('   Los 5 productos con mayor participación son:')
    display(
        df[['Producto', 'Total_Venta', 'Porcentaje_Ventas']]
        .sort_values('Porcentaje_Ventas', ascending=False)
        .head(5)
        .assign(Porcentaje_Ventas=lambda x: x['Porcentaje_Ventas'].map('{:.2f}%'.format))
        .reset_index(drop=True)
    )
else:
    display(
        productos_top[['Producto', 'Total_Venta', 'Porcentaje_Ventas']]
        .sort_values('Porcentaje_Ventas', ascending=False)
        .assign(Porcentaje_Ventas=lambda x: x['Porcentaje_Ventas'].map('{:.2f}%'.format))
        .reset_index(drop=True)
    )
```

---
## 📌 Punto 4 — Estadística Descriptiva

```python
# ── 4.1 Media, Mediana y Moda del Precio ─────────────────────────────────────
media_precio   = df['Precio'].mean()
mediana_precio = df['Precio'].median()
moda_precio    = df['Precio'].mode()

print('=' * 60)
print('📊 ESTADÍSTICAS DE LA COLUMNA PRECIO')
print('=' * 60)
print(f'   • Media   : $ {media_precio:,.0f}')
print(f'   • Mediana : $ {mediana_precio:,.0f}')
print(f'   • Moda    : {moda_precio.values}')
print()
print('💡 Interpretación:')
print(f'   → Media ({media_precio:,.0f}): Precio promedio de todos los productos.')
print(f'   → Mediana ({mediana_precio:,.0f}): El 50% de los productos cuesta menos de este valor.')
print(f'   → Moda: El/los precio(s) que más se repite(n) en el catálogo.')
```

```python
# ── 4.2 Desviación estándar de Cantidad ──────────────────────────────────────
std_cantidad = df['Cantidad'].std()

print('=' * 60)
print('📊 DESVIACIÓN ESTÁNDAR DE CANTIDAD')
print('=' * 60)
print(f'   • Desviación estándar: {std_cantidad:.2f} unidades')
print(f'   • Media de Cantidad  : {df["Cantidad"].mean():.2f} unidades')
print()
print('💡 Interpretación:')
print(f'   La desviación de {std_cantidad:.2f} indica variabilidad en el stock.')
print(f'   Algunos productos tienen muchas unidades y otros pocas.')
```

```python
# ── 4.3 Máximo y mínimo de Total_Venta ───────────────────────────────────────
max_tv   = df['Total_Venta'].max()
min_tv   = df['Total_Venta'].min()
rango    = max_tv - min_tv
prod_max = df.loc[df['Total_Venta'].idxmax(), 'Producto']
prod_min = df.loc[df['Total_Venta'].idxmin(), 'Producto']

print('=' * 60)
print('📊 RANGO DE TOTAL_VENTA')
print('=' * 60)
print(f'   • Máximo : $ {max_tv:,.0f}  → {prod_max}')
print(f'   • Mínimo : $ {min_tv:,.0f}  → {prod_min}')
print(f'   • Rango  : $ {rango:,.0f}')
print()
print('💡 Interpretación:')
print(f'   El rango de $ {rango:,.0f} indica alta dispersión en el')
print(f'   desempeño de ventas entre productos.')
```
